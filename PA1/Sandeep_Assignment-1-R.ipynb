```r
#1-d Find all percentage of 'Residence_type' for all the values
dataframe %>% 
    group_by(Residence_type) %>% 
        summarise(percent=100*n()/nrow(dataframe))
```

```r
# Task 2: Aggregation & Filtering & Rank
```

```r
#Task 2-a: Find out the gender with largest number of records 
dfgender <- dataframe %>% count(gender=dataframe$gender)
#dfgender %>% summarize(records=max(dfgender$n))
dfgender
# from the output 
```

```r
#Task 2-b: Find out the total number of Residence_type "Urban" who are Male
filterdf <- dataframe %>% filter(Residence_type=='Urban' & gender=='Male')
count(filterdf) 
```

```r
# Group by function for dataframe in R using pipe operator 
#2-c 1 question #Find the top 10 ages with highest av_glucose_level
df1 <- dataframe %>% 
        group_by(age) %>%
        summarise(glucose_level=max(avg_glucose_level))
df1[order(-df1$age),][1:10,]
```

```r
#2-d 2nd question top 10 ages with more number of strokes
df2<- dataframe %>% group_by(age) %>% summarise(high_stroke=max(stroke))
df2[order(-df2$age),][1:10,]
```

```r
##TASK 3: VISUALIZATION
```

```r
#task 3-a
#Create barplot showing gender with count with residence type
df_3 <- dataframe %>% select(Residence_type,gender)
df_3_1<- table(df_3)
df_3_1
barplot(df_3_1,main='Gender Vs Residence Type',xlab="Gender",col=c("Green","Yellow"),legend=rownames(df_3_1))
```

```r
#task 3-b
#Display pie chart for the smoking status data
df4_1 <- dataframe %>% group_by(smoking_status) %>% filter(smoking_status=='never smoked' | smoking_status=='smokes'|
                                                          smoking_status=='formerly smoked')
df4_1_1<-table(df4_1['smoking_status']) 
lbls<- c("formerly smoked","never smoked","smokes")
pct=round(df4_1_1/sum(df4_1_1)*100)
new_labels=paste(lbls,"-",pct,"%",sep="")
pie(df4_1_1,labels=new_labels,main="Smoking status data")
```

```r
#Task4 finding an interesting pattern
# atleast two visualization with explanation
#First visualization 
df5_1 <- dataframe %>% group_by(Residence_type) %>% filter(Residence_type=='Urban' | Residence_type=='Rural')
df5_1_1<-table(df4_1['Residence_type']) 
lbls<- c("Rural","Urban")
pct=round(df5_1_1/sum(df5_1_1)*100)
new_labels=paste(lbls,"-",pct,"%",sep="")
pie(df5_1_1,labels=new_labels,main="Residence status data")


#Second Visualization
```
